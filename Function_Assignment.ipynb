

## 1. Difference between a Function and a Method in Python

**Function:** a function is a block of code that can perform some task basically it helps to do tasks more efficiently and fast.

**Method:** A function that is associated with an object and is called on that object.

```python
# Function
def greet(name):
    return f"Hello, {name}!"

print(greet("Aakash"))  # Output: Hello, Aakash!

```



## 2. Function Arguments and Parameters in Python

**Parameter:** A variable defined in the function to accept input.
**Argument:** The actual value passed to the function when called.

```python
def add(a, b):  # a and b are parameters
    return a + b

print(add(5, 3))  # 5 and 3 are arguments, Output: 8
```

## 3. Different Ways to Define and Call a Function

* Normal function with positional arguments
* Function with default arguments
* Function with keyword arguments
* Function with arbitrary arguments (`*args`, `**kwargs`)

```python
# Normal function
def greet(name):
    print(f"Hello, {name}!")

greet("Aakash")  # Positional argument

# Default argument
def greet(name="Friend"):
    print(f"Hello, {name}!")

greet()  # Output: Hello, Friend!

# Keyword arguments
def greet(name, age):
    print(f"Hello, {name}. You are {age} years old.")

greet(age=22, name="Aakash")


```

## 4. Purpose of the `return` Statement

* The return statement is used to send a value back to the caller from a function note without `return`, the function returns `None`.

```python
def square(x):
    return x ** 2

```


## 5. Iterators vs Iterables in Python

* **Iterable:** An object that can return an iterator (e.g., list, tuple, string)
* **Iterator:** An object that keeps state and produces the next value when `next()` is called

```python
# Iterable
my_list = [1, 2, 3]

# Iterator
my_iter = iter(my_list)

print(next(my_iter))  # 1
print(next(my_iter))  # 2
print(next(my_iter))  # 3
# print(next(my_iter))  #error
```

---

## 6. Generators in Python

* **Generator:** A special type of iterator that generates values on the fly, using yield instead of return.
* **Defined using `yield` keyword**

```python
def my_generator():
    for i in range(3):
        yield i

gen = my_generator()
print(next(gen))  # 0
print(next(gen))  # 1
print(next(gen))  # 2
```


## 7. Advantages of Using Generators

* **Memory efficient:** Generates items one at a time
* **Lazy evaluation:** Useful for large datasets
* **Can represent infinite sequences**


## 8. Lambda Function in Python

* **Anonymous (unnamed) function** defined using `lambda`
* can be easily use with map , filter and reduce
```python
square = lambda x: x**2
print(square(5))  # Output: 25
```


## 9. Purpose and Usage of `map()`

* `map()` applies a function to **all items in an iterable** and returns a map object

```python
numbers = [1, 2, 3, 4]
squared = map(lambda x: x**2, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]
```


## 10. Difference between `map()`, `reduce()`, and `filter()`

map() → “Change everything”
Imagine you have a box of apples, and you paint every apple red. map() changes each item in the box.

filter() → “Pick only some”
You have a basket of fruits, and you take only the apples. filter() keeps only the things you want.

reduce() → “Combine everything”
You have coins, and you add all the coins together to see the total. reduce() combines all items into one single result.

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even_numbers(nums))
````

````python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(numbers):
    return [num ** 2 for num in numbers]

nums = [1, 2, 3, 4]
print("Squares of numbers:", square_list(nums))
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
print("Primes from 1 to 200:")
for i in range(1, 201):
    if is_prime(i):
        print(i, end=' ')
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fib:
    def __init__(self, n):
        self.n, self.a, self.b = n, 0, 1

    def __iter__(self): return self

    def __next__(self):
        if self.n == 0: raise StopIteration
        self.n -= 1
        self.a, self.b = self.b, self.a + self.b
        return self.a

for i in Fib(15):
    print(i, end=",")

````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for x in powers_of_two(5):
    print(x)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
````

````python

# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 3), (4, 1), (2, 2)]

data.sort(key=lambda x: x[1])
print(data)
````

````python

# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius = [0, 10, 30, 300]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

text = "Aakash Bunny"
vowels = "aeiouAEIOU"

result = "".join(filter(lambda x: x not in vowels, text))
print(result)
````
