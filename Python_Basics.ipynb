#Python Basics


1. What is Python, and why is it popular ?

   => Python is a high-level programming language known for its simple,       readable syntax. Its popular because its easy to learn, works on many platforms, and has powerful libraries for things like web development, data science, AI, ML, and automation.


2. What is an interpreter in Python ?
  
   => An interpreter in Python is the program that reads and runs Python code line by line. It translates your code into machine-readable instructions so the computer can execute it.


3. What are pre-defined keywords in Python ?

   => Pre-defined keywords in Python are special reserved words that have fixed meanings and are used to define the structure of the language. You can not use them as variable names because Python already uses them for specific functions, like if, while, for, class, etc.


4. Can keywords be used as variable names ?

   => No, keywords can't be used as variable names because they already have special meanings in Python and are part of the language's syntax.


5. What is mutability in Python ?

   => Mutability in Python means whether a value can be changed after it's created.Mutable objects like lists and dictionaries can be updated, while immutable ones like strings and tuples cannot be changed once made.


6. Why are lists mutable, but tuples are immutable ?

   => Lists are designed to allow changes—like adding, removing, or modifying items—so they're mutable. Tuples, on the other hand, are meant to store fixed data, so Python makes them immutable, meaning their values can't be changed once created.


7. What is the difference between “==” and “is” operators in Python ?

   => '==' checks if two values are equal, while 'is' checks if they refer to the exact same object in memory. So, two things can be equal in value but still not be the same object.

  
8. What are logical operators in Python ?

   => Logical operators in Python are used to combine or compare conditions.
   They are **and**, **or**, and **not**, and they help decide whether a condition is true or false.


9. What is type casting in Python ?

   => Type casting in Python means converting a value from one data type to another, like turning a string into an integer or a float into a string. It helps when you need values in a specific format to perform operations.


10. What is the difference between implicit and explicit type casting ?

    => Implicit type casting happens automatically when Python converts one data type to another without you asking for it—usually when it's safe, like converting an int to a float during calculations.
    Explicit type casting is when you manually convert a value using functions like int(), float(), or str() to change its type on purpose.


11. What is the purpose of conditional statements in Python ?

    => Conditional statements let your program make decisions. They check conditions and run certain blocks of code only when those conditions are true, helping control the flow of the program.


12. How does the elif statement work ?

    => The 'elif' statement checks another condition if the previous 'if' condition was false. If the 'elif' condition is true, its block runs; if not, Python moves on to the next 'elif' or 'else'.

13. What is the difference between for and while loops ?

    => A 'for loop' is used when you know how many times you want to repeat something or when you're looping through items in a sequence.
    A 'while loop' repeats an action until a certain condition becomes false, so it depends on the condition rather than a fixed count.


14. Describe a scenario where a while loop is more suitable than a for loop.

    => A 'while loop' is more suitable when you don’t know how many times the loop needs to run. For example, if you're asking a user to enter a password and you want to keep asking until they type the correct one, a while loop works better because the number of attempts isn't fixed.

```python
# 5. Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

# Creating a list (mutable)
my_list = [1, 2, 3]

# Creating a tuple (immutable)
my_tuple = (1, 2, 3)

# Changing an element in the list - this works
my_list[1] = 10
print("Updated list:", my_list)

# Trying to change an element in the tuple - this will cause an error
try:
    my_tuple[1] = 10
except TypeError:
    print("Tuples cannot be changed because they are immutable.")
```

```python
# Immutable example (number)
def change_num(n):
    n = n + 1
    print("Inside function:", n)

# Mutable example (list)
def change_list(lst):
    lst.append(100)
    print("Inside function:", lst)

num = 5
my_list = [1, 2, 3]

change_num(num)
print("Outside function:", num)     # number does NOT change

change_list(my_list)
print("Outside function:", my_list) # list DOES change
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

total = 0

for i in range(1, 51):
    if i % 2 == 0:      # check even
        total += i

print("Sum:", total)

```

```python
# 14. Write a program to reverse a string using a while loop.

text = input("Enter text: ")
rev = ""
i = len(text) - 1      # start from last index

while i >= 0:
    rev += text[i]
    i -= 1

print("Reversed:", rev)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i          # multiply each number
    i += 1

print("Factorial:", fact)
```
